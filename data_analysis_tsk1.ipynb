```python
import matplotlib.pyplot as plt
np.random.seed(1)
values = np.random.randn(1000).cumsum()

plt.plot(range(0,1000),values)
plt.show()
```

```python
max_drawdown = np.max(np.maximum.accumulate(values) - values)
print(max_drawdown)
```
